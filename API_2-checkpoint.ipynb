```python
# trying to get all comments with another approch
import requests

post_id = "C3YRsjQvG0i"
api_endpoint = f"https://api.data365.co/v1.1/instagram/post/{post_id}/comments"
access_token = "ZXlKMGVYQWlPaUpLVjFRaUxDSmhiR2NpT2lKSVV6STFOaUo5LmV5SnpkV0lpT2lKUVpYSnBlbVZ5SWl3aWFXRjBJam94TnpFMk9Ea3pNelV6TGpnMU16azFOeko5LmZoSXFVeGRRWnJJS0ZhRVRGSm9LN1BiUjJBTlpNeUNuTlJiVm5jR04yQm8="

# Set parameters for fetching comments
params = {
    "access_token": access_token,
    # "max_page_size": 100,  # Set the maximal number of items per page to 100
}

# Initialize a list to store all comments
all_comments = []

# Make the GET request to fetch comments for the post with pagination
cursor = None
while True:
    if cursor:
        params['cursor'] = cursor

    response = requests.get(api_endpoint, params=params)
    

    # Check the response
    if response.status_code == 200:
        data = response.json()
        # Append comments to the list
        all_comments.extend(data['data']['items'])

        # Check if there are more pages
        if not data['data']['page_info']['has_next_page']:
            break  # Exit the loop if there are no more pages

        # Update the cursor for the next page
        cursor = data['data']['page_info']['cursor']
    else:
        print("Error:", response.status_code)
        print(response.text)
        break

# Process all comments
for comment in all_comments:
    print(comment)
```

```python
# trying for other post id
import requests

def fetch_all_comments(post_id, access_token):
    api_endpoint = f"https://api.data365.co/v1.1/instagram/post/{post_id}/comments"
    
    # Set initial parameters for fetching comments
    params = {
        "access_token": access_token,
        "max_page_size": 100  # Set the maximal number of items per page to 100
    }

    all_comments = []
    cursor = None

    while True:
        if cursor:
            params['cursor'] = cursor
        
        response = requests.get(api_endpoint, params=params)

        if response.status_code == 200:
            data = response.json()
            # Append comments to the list
            all_comments.extend(data['data']['items'])
            
            # Check if there are more pages
            if not data['data']['page_info']['has_next_page']:
                break  # Exit the loop if there are no more pages

            # Update the cursor for the next page
            cursor = data['data']['page_info']['cursor']
        else:
            print("Error:", response.status_code)
            print(response.text)
            break

    return all_comments

# Example usage
post_id = "C6q0b0fPF23"
access_token = "ZXlKMGVYQWlPaUpLVjFRaUxDSmhiR2NpT2lKSVV6STFOaUo5LmV5SnpkV0lpT2lKUVpYSnBlbVZ5SWl3aWFXRjBJam94TnpFMk9Ea3pNelV6TGpnMU16azFOeko5LmZoSXFVeGRRWnJJS0ZhRVRGSm9LN1BiUjJBTlpNeUNuTlJiVm5jR04yQm8="

comments = fetch_all_comments(post_id, access_token)
print(comments)
```
